# **APRIORI**

```python
from pyECLAT import Example2
dados_example02 = Example2().get()
dados_example02
```

```python
dados_example02.shape
```

```python
from pyECLAT import ECLAT
eclat = ECLAT(data=dados_example02)
dados_example02_bin = eclat.df_bin
dados_example02_bin
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Gerando a associação
associacao_apriori = apriori(dados_example02_bin, min_support=0.03, use_colnames=True)
```

```python
# Colocando em ordem dos mais frequentes
associacao_apriori.sort_values("support", ascending=False).head(15)
```

```python
# Criando as regras de associação
regras = association_rules(associacao_apriori, metric="lift")
```

```python
# Colocando em ordem de suporte, confiança ou grau de associação (Lift)
resultado = regras.sort_values("lift", ascending=False)
```
